```python
label_path = "/content/drive/MyDrive/Dissertation_Project/data/raw/anomaly_labels"
label_contents = os.listdir(label_path)
label_path = "/content/drive/MyDrive/Dissertation_Project/data/raw/anomaly_labels"

# Dictionary to store loaded label arrays
label_data = {}

# Iterate through each file in the label directory
for file_name in os.listdir(label_path):
    if file_name.endswith('.npy'):
        file_path = os.path.join(label_path, file_name)

        # Load the label array from the .npy file
        label_array = np.load(file_path)

        # Store the loaded label array in the dictionary
        label_data[file_name] = label_array

# Print the dictionary containing the loaded label arrays
print("Loaded label data:")
for file_name, label_array in label_data.items():
    print(f"File: {file_name}, Label Array: {label_array}")
```
